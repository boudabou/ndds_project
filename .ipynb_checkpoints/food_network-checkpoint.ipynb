## 1.3 Cleaning the dataset :

```python
#A list to keep track of values that couldn't be converted to a float
special_values=[]

def cleanup(food, columns) : 
    """
    Returns vector with columns provided in parameter columns and float values.
    :param food: list
    :param columns: list of strings
    :return: DataFrame with float values
    """
    vector = food[columns] 
    for i in columns : 
        #Skip columns with text.
        if(i in ['ORIGFDNM','ORIGGPFR']) : continue
        #convert string values to float.
        try : 
            vector[i]=float(vector[i])
        #If that is not possible set it to 0.
        except ValueError : 
            #Kepping track of characters that will be assigned 0.
            special_values.append(vector[i])
            vector[i]=0
    return vector
```

```python
#Selecting columns that we will keep for our graph. 

columns= ['ORIGFDNM','ORIGGPFR','ORIGGPCD', #These columns (ORIGFDNM and ORIGGPFR) are kept for later use. But they won't affect the computations to come.
          'Energie, Règlement UE N° 1169/2011 (kcal/100g)',
          'Protéines (g/100g)','Glucides (g/100g)',
          'Fibres alimentaires (g/100g)','Lipides (g/100g)',
          'AG saturés (g/100g)','Cholestérol (mg/100g)',
          'Alcool (g/100g)',
          'Sel chlorure de sodium (g/100g)',
          'Calcium (mg/100g)',
          'Chlorure (mg/100g)',
          'Cuivre (mg/100g)',
          'Fer (mg/100g)',
          'Iode (µg/100g)',
          'Magnésium (mg/100g)',
          'Manganèse (mg/100g)',
          'Phosphore (mg/100g)',
          'Potassium (mg/100g)',
          'Sélénium (µg/100g)',
          'Sodium (mg/100g)',
          'Zinc (mg/100g)',
          'Rétinol (µg/100g)',
          'Beta-Carotène (µg/100g)',
          'Vitamine D (µg/100g)',
          'Vitamine E (mg/100g)',
          'Vitamine K1 (µg/100g)',
          'Vitamine K2 (µg/100g)',
          'Vitamine C (mg/100g)',
          'Vitamine B1 ou Thiamine (mg/100g)',
          'Vitamine B2 ou Riboflavine (mg/100g)',
          'Vitamine B3 ou PP ou Niacine (mg/100g)',
          'Vitamine B5 ou Acide pantothénique (mg/100g)',
          'Vitamine B6 (mg/100g)',
          'Vitamine B9 ou Folates totaux (µg/100g)',
          'Vitamine B12 (µg/100g)']
```

# 2. Constructing food graph : 

Foods will be represented as n-dimentional vectors of the attributes of the food (calories, protein, carbs and so on...). Therefore, the edges will be the distance between the vectors (foods) and the nodes will be the foods themselves.

```python
def return_node_list(dataframe) : 
    nodes='{\n "nodes": [\n'
    first_index = dataframe.index[0]
    for food_index in dataframe.index:
        food=dataframe.loc[food_index]
        
        nodes += '\t{"id":'+ str(food_index) + ', "label": "'+ dataframe['ORIGFDNM'][food_index].replace('"',"'")+'", "group":'+str(dataframe['ORIGGPCD'][food_index]).split('.')[1]
        if (food_index==max(dataframe.index)) : nodes +='}\n'
        else : nodes +='},\n'
    nodes += '],\n'
    return nodes,first_index
```

```python
vectors['group'] = vectors['ORIGGPCD'].apply(lambda x : int(str(x).split('.')[0]))
groups = []
n_groups = len(np.unique(vectors['group']))
for i in range(n_groups) : #Number of groups.
    groups.append(vectors[vectors['group']==i+1])
```

```python
nodes_list = []
first_indices = []
for i in range(n_groups) : 
    nodes,index = return_node_list(groups[i])
    nodes_list.append(nodes)
    first_indices.append(index)
```

```python
import scipy.spatial.distance
def compute_distances(dataframe) : 
    
    distances = scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(dataframe.values,p=2,metric='cosine'))
    return distances
```

```python
features = [groups[i][columns[3:]].copy() for i in range(n_groups)]
distances=[]
for i in range(n_groups) :
    features[i] -= features[i].mean(axis=0)
    features[i] /= features[i].std(axis=0)
    features[i]=features[i].fillna(0.0)
    distances.append(compute_distances(features[i]))
```

```python
def create_edge_list(distances,vectors,n_neighbors,first_index):
    edge_list='\t"edges": [\n'
    
    for dist in range(len(distances)):
        node=distances[dist]
        
        indices = np.argsort(node)[1:n_neighbors+1]
        
        for i in range(len(indices)):
            
            if(dist==len(distances)-1 and i == len(indices)-1) :  
                edge_list+='\t\t{"from": '+str(first_index+dist)+', "to": '+str(first_index+indices[i])+'}\n'
            edge_list+='\t\t{"from": '+str(first_index+dist)+', "to": '+str(first_index+indices[i])+'},\n'
    edge_list+=' ]\n}'
    return edge_list
```

```python
ls=[create_edge_list(distances[i],groups[i],10,first_indices[i]) for i in range(n_groups)]
```

```python
json_file=[]
[json_file.append(nodes_list[i]+ls[i]) for i in range(n_groups)]

#write file in current directory
for i in range(n_groups) : 
    file = open('graphTest'+str(i)+'.json','w')
    d = file.write(json_file[i])
    file.close()
```

```python
#Reading the graph.
food_graph = nx.read_edgelist('edgelist',nodetype=int,data=(('weight',float),))
```

It is useless to keep a connection between all nodes. So we will delete the edges that have weights that are too big. 

```python
def delete_edges(graph,thres) : 
    """
    Returns the graph after deleting the edges with weights higher than thres.
    :param graph: Graph
    :param thres: float
    :return: graph
    """
    for i in graph.edges(data=True) : 
        if(i[2]['weight']>thres) : graph.remove_edge(i[0],i[1])
    return graph
```

Let's take a threshold of thres=300

```python
#The value of the threshold is to be reevaluated.
new_graph=delete_edges(food_graph,300)
```

# 3. Drawing the graph :
